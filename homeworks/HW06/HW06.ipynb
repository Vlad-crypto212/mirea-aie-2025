```python
import os
import pandas as pd
import numpy as np
import joblib
import json

# Модели и инструменты sklearn
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, StackingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.inspection import permutation_importance
from sklearn.datasets import make_classification # Для генерации примера, если нужно

# Визуализация
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_path = 'S06-hw-dataset-02.csv'
df = pd.read_csv(file_path)

print("--- Head данных ---")
display(df.head()) # Используем display для красивого вывода в Jupyter

print("\n--- Info о данных ---")
display(df.info())

print("\n--- Основные статистики ---")
display(df.describe())

print("\n--- Распределение таргета (target) ---")
target_dist = df['target'].value_counts(normalize=True)
print(target_dist)
sns.countplot(data=df, x='target')
plt.title('Распределение классов в target')
plt.savefig('artifacts/figures/target_distribution.png')
plt.show()

print("\n--- Проверка пропусков ---")
print(df.isnull().sum().sum()) # Суммарное количество пропусков

# Определение X и y
X = df.drop(columns=['target'])
y = df['target']

print(f"\nКоличество признаков (X): {X.shape[1]}")
print(f"Количество образцов: {X.shape[0]}")
print(f"Целевая переменная (y): {y.name}")

print("\n--- 2.3.1. Загрузка данных и первичный анализ завершена ---")
```

```python
RANDOM_STATE = 42
TEST_SIZE = 0.2

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)

print(f"Размер X_train: {X_train.shape}")
print(f"Размер X_test: {X_test.shape}")
print(f"Размер y_train: {y_train.shape}")
print(f"Размер y_test: {y_test.shape}")

# Проверка стратификации
print("\n--- Баланс классов в y_train ---")
print(y_train.value_counts(normalize=True))
print("\n--- Баланс классов в y_test ---")
print(y_test.value_counts(normalize=True))

print("\n--- Пояснение ---")
print("""
Фиксированный random_state (например, 42) гарантирует, что при каждом запуске разбиения
мы получим одинаковые тренировочные и тестовые наборы. Это важно для воспроизводимости
экспериментов и честного сравнения моделей.

Stratify=y означает, что соотношение классов в target (y) будет примерно одинаковым
в тренировочной и тестовой выборках. Это особенно критично для задач классификации,
чтобы ни одна из выборок не была смещенной и не содержала, например,
только один класс.
""")

print("\n--- 2.3.2. Train/Test-сплит завершён ---")
```

```python
print("--- Обучение Baseline моделей ---")

# 1. Dummy Classifier
dummy_clf = DummyClassifier(strategy='most_frequent', random_state=RANDOM_STATE)
dummy_clf.fit(X_train, y_train)
y_pred_dummy = dummy_clf.predict(X_test)

acc_dummy = accuracy_score(y_test, y_pred_dummy)
f1_dummy = f1_score(y_test, y_pred_dummy)
# ROC-AUC может быть не определён для most_frequent, если в тесте нет положительных примеров
try:
    y_proba_dummy = dummy_clf.predict_proba(X_test)[:, 1] # Вероятности класса 1
    auc_dummy = roc_auc_score(y_test, y_proba_dummy)
except ValueError as e:
    print(f"ROC-AUC для DummyClassifier не рассчитан: {e}")
    auc_dummy = 0.5 # Для равновзвешенного случая

print(f"Baseline (Dummy - most_frequent): Accuracy={acc_dummy:.4f}, F1={f1_dummy:.4f}, AUC={auc_dummy:.4f}")


# 2. Logistic Regression (Pipeline)
lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('log_reg', LogisticRegression(random_state=RANDOM_STATE, max_iter=1000))
])

lr_pipeline.fit(X_train, y_train)
y_pred_lr = lr_pipeline.predict(X_test)
y_proba_lr = lr_pipeline.predict_proba(X_test)[:, 1] # Для AUC

acc_lr = accuracy_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)
auc_lr = roc_auc_score(y_test, y_proba_lr)

print(f"Baseline (Logistic Regression): Accuracy={acc_lr:.4f}, F1={f1_lr:.4f}, AUC={auc_lr:.4f}")

# Сохранение результатов baseline'ов для финального отчета
baseline_results = {
    'Dummy_Most_Frequent': {'accuracy': acc_dummy, 'f1': f1_dummy, 'roc_auc': auc_dummy},
    'Logistic_Regression': {'accuracy': acc_lr, 'f1': f1_lr, 'roc_auc': auc_lr}
}

print("\n--- Baseline'ы обучены и оценены ---")
```

```python
print("--- Обучение Decision Tree ---")
dt_clf = DecisionTreeClassifier(random_state=RANDOM_STATE)

# Подбор гиперпараметров
param_grid_dt = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_dt = GridSearchCV(dt_clf, param_grid_dt, cv=5, scoring='roc_auc', n_jobs=-1)
grid_search_dt.fit(X_train, y_train)

best_dt = grid_search_dt.best_estimator_
print(f"Лучшие параметры для DT: {grid_search_dt.best_params_}")

y_pred_dt = best_dt.predict(X_test)
y_proba_dt = best_dt.predict_proba(X_test)[:, 1]

acc_dt = accuracy_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
auc_dt = roc_auc_score(y_test, y_proba_dt)

print(f"Decision Tree (Best): Accuracy={acc_dt:.4f}, F1={f1_dt:.4f}, AUC={auc_dt:.4f}")

# Сохранение результатов DT
dt_results = {
    'model_type': 'DecisionTree',
    'best_params': grid_search_dt.best_params_,
    'cv_score': grid_search_dt.best_score_,
    'test_metrics': {'accuracy': acc_dt, 'f1': f1_dt, 'roc_auc': auc_dt}
}
```

```python
print("\n--- Обучение Random Forest ---")
rf_clf = RandomForestClassifier(random_state=RANDOM_STATE)

# Упрощённый сет подбора параметров для ускорения
param_grid_rf = {
    'n_estimators': [50, 100], # Уменьшено количество вариантов
    'max_depth': [5, 10, None], # Уменьшено количество вариантов
    'min_samples_split': [2, 10], # Уменьшено количество вариантов
    # 'min_samples_leaf': [1, 2] # Закомментировано для ускорения, можно добавить позже
}

# Уменьшено количество фолдов CV до 3 для ускорения
grid_search_rf = GridSearchCV(rf_clf, param_grid_rf, cv=3, scoring='roc_auc', n_jobs=-1, verbose=1) 
grid_search_rf.fit(X_train, y_train)

best_rf = grid_search_rf.best_estimator_
print(f"Лучшие параметры для RF: {grid_search_rf.best_params_}")

y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)[:, 1]

acc_rf = accuracy_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)
auc_rf = roc_auc_score(y_test, y_proba_rf)

print(f"Random Forest (Best): Accuracy={acc_rf:.4f}, F1={f1_rf:.4f}, AUC={auc_rf:.4f}")

# Сохранение результатов RF
rf_results = {
    'model_type': 'RandomForest',
    'best_params': grid_search_rf.best_params_,
    'cv_score': grid_search_rf.best_score_,
    'test_metrics': {'accuracy': acc_rf, 'f1': f1_rf, 'roc_auc': auc_rf}
}
```

```python
print("\n--- Обучение Gradient Boosting ---")
gb_clf = GradientBoostingClassifier(random_state=RANDOM_STATE)

# Упрощённый сет подбора параметров для ускорения
param_grid_gb = {
    'n_estimators': [50, 100], # Уменьшено количество вариантов
    'learning_rate': [0.1, 0.2], # Уменьшено количество вариантов
    'max_depth': [3, 5] # Уменьшено количество вариантов
}

# Уменьшено количество фолдов CV до 3 для ускорения
grid_search_gb = GridSearchCV(gb_clf, param_grid_gb, cv=3, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_search_gb.fit(X_train, y_train)

best_gb = grid_search_gb.best_estimator_
print(f"Лучшие параметры для GB: {grid_search_gb.best_params_}")

y_pred_gb = best_gb.predict(X_test)
y_proba_gb = best_gb.predict_proba(X_test)[:, 1]

acc_gb = accuracy_score(y_test, y_pred_gb)
f1_gb = f1_score(y_test, y_pred_gb)
auc_gb = roc_auc_score(y_test, y_proba_gb)

print(f"Gradient Boosting (Best): Accuracy={acc_gb:.4f}, F1={f1_gb:.4f}, AUC={auc_gb:.4f}")

# Сохранение результатов GB
gb_results = {
    'model_type': 'GradientBoosting',
    'best_params': grid_search_gb.best_params_,
    'cv_score': grid_search_gb.best_score_,
    'test_metrics': {'accuracy': acc_gb, 'f1': f1_gb, 'roc_auc': auc_gb}
}
```

```python
print("\n--- Обучение StackingClassifier ---")
base_models = [
    ('lr', Pipeline([('scaler', StandardScaler()), ('log_reg', LogisticRegression(random_state=RANDOM_STATE, max_iter=1000))]) ),
    ('dt', DecisionTreeClassifier(max_depth=5, random_state=RANDOM_STATE)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE))
]
meta_model = LogisticRegression(random_state=RANDOM_STATE)

stacking_clf = StackingClassifier(estimators=base_models, final_estimator=meta_model, cv=5)
stacking_clf.fit(X_train, y_train) # StackingClassifier сам делает CV для базовых моделей

y_pred_stacking = stacking_clf.predict(X_test)
y_proba_stacking = stacking_clf.predict_proba(X_test)[:, 1]

acc_stacking = accuracy_score(y_test, y_pred_stacking)
f1_stacking = f1_score(y_test, y_pred_stacking)
auc_stacking = roc_auc_score(y_test, y_proba_stacking)

print(f"StackingClassifier: Accuracy={acc_stacking:.4f}, F1={f1_stacking:.4f}, AUC={auc_stacking:.4f}")

# Сохранение результатов Stacking
stacking_results = {
    'model_type': 'StackingClassifier',
    'best_params': 'N/A (uses default or pre-defined)', # Параметры базовых моделей уже фиксированы или определены через CV
    'cv_score': 'N/A (calculated internally)', # Внутренняя логика StackingClassifier
    'test_metrics': {'accuracy': acc_stacking, 'f1': f1_stacking, 'roc_auc': auc_stacking}
}
```

```python

all_results = {
    'Dummy_Most_Frequent': baseline_results['Dummy_Most_Frequent'],
    'Logistic_Regression': baseline_results['Logistic_Regression'],
    'Decision_Tree': dt_results['test_metrics'],
    'Random_Forest': rf_results['test_metrics'],
    'Gradient_Boosting': gb_results['test_metrics'],
    # Проверяем, были ли результаты для Stacking (он опциональный)
    'Stacking_Classifier': stacking_results.get('test_metrics', {'accuracy': 0, 'f1': 0, 'roc_auc': 0}) if 'stacking_results' in locals() else {'accuracy': 0, 'f1': 0, 'roc_auc': 0}
}

results_df = pd.DataFrame(all_results).T
print("--- Сравнение метрик моделей на тесте ---")
display(results_df)

# Выбор лучшей модели по AUC (для бинарной классификации часто используется AUC)
best_model_name = results_df['roc_auc'].idxmax()
best_model_auc_value = results_df.loc[best_model_name, 'roc_auc']
print(f"\nЛучшая модель по ROC-AUC: {best_model_name}")
print(f"AUC лучшей модели: {best_model_auc_value:.4f}")

# --- Исправленный способ получения объекта лучшей модели ---
print(f"\n--- Определение объекта лучшей модели: {best_model_name} ---")

# Инициализируем best_model_instance как None
best_model_instance = None

# Сопоставляем имя лучшей модели с соответствующим объектом
if best_model_name == 'Decision_Tree':
    best_model_instance = best_dt
    print(f"Выбрана модель: Decision Tree (best_dt)")
elif best_model_name == 'Random_Forest':
    best_model_instance = best_rf
    print(f"Выбрана модель: Random Forest (best_rf)")
elif best_model_name == 'Gradient_Boosting':
    best_model_instance = best_gb
    print(f"Выбрана модель: Gradient Boosting (best_gb)")
elif best_model_name == 'Logistic_Regression':
    best_model_instance = lr_pipeline
    print(f"Выбрана модель: Logistic Regression Pipeline (lr_pipeline)")
elif best_model_name == 'Stacking_Classifier':
    # Убедимся, что переменная stacking_clf существует
    if 'stacking_clf' in locals():
        best_model_instance = stacking_clf
        print(f"Выбрана модель: Stacking Classifier (stacking_clf)")
    else:
        print("ПРЕДУПРЕЖДЕНИЕ: Переменная 'stacking_clf' не найдена, невозможно выбрать модель Stacking.")
        # Можно добавить логику обработки ошибки или выбрать другую модель
        # Например, выбрать вторую по значению AUC модель
        second_best_name = results_df.nlargest(2, 'roc_auc').index[-1]
        print(f"Попробуем использовать вторую лучшую модель: {second_best_name}")
        # Повторяем логику для second_best_name, если нужно
        # Лучше всего обернуть логику выбора модели в функцию, чтобы избежать дублирования кода
        # Но для простоты вставим сюда нужные elif'ы для second_best_name
        if second_best_name == 'Decision_Tree':
            best_model_instance = best_dt
            best_model_name = second_best_name # Обновляем имя для последующих ячеек
            print(f"Выбрана модель: Decision Tree (best_dt) как альтернатива.")
        elif second_best_name == 'Random_Forest':
            best_model_instance = best_rf
            best_model_name = second_best_name
            print(f"Выбрана модель: Random Forest (best_rf) как альтернатива.")
        elif second_best_name == 'Gradient_Boosting':
            best_model_instance = best_gb
            best_model_name = second_best_name
            print(f"Выбрана модель: Gradient Boosting (best_gb) как альтернатива.")
        elif second_best_name == 'Logistic_Regression':
            best_model_instance = lr_pipeline
            best_model_name = second_best_name
            print(f"Выбрана модель: Logistic Regression Pipeline (lr_pipeline) как альтернатива.")
        else:
             print(f"ПРЕДУПРЕЖДЕНИЕ: Ни одна из известных моделей не была найдена для {second_best_name}. best_model_instance останется None.")
elif best_model_name == 'Dummy_Most_Frequent':
    # DummyClassifier не подходит для Permutation Importance, как указано в задании
    print(f"ПРЕДУПРЕЖДЕНИЕ: DummyClassifier не подходит для интерпретации Permutation Importance.")
    # Выберем вторую по значению AUC модель
    second_best_name = results_df.nlargest(2, 'roc_auc').index[-1]
    print(f"Попробуем использовать вторую лучшую модель: {second_best_name}")
    if second_best_name == 'Decision_Tree':
        best_model_instance = best_dt
        best_model_name = second_best_name
        print(f"Выбрана модель: Decision Tree (best_dt) как альтернатива.")
    elif second_best_name == 'Random_Forest':
        best_model_instance = best_rf
        best_model_name = second_best_name
        print(f"Выбрана модель: Random Forest (best_rf) как альтернатива.")
    elif second_best_name == 'Gradient_Boosting':
        best_model_instance = best_gb
        best_model_name = second_best_name
        print(f"Выбрана модель: Gradient Boosting (best_gb) как альтернатива.")
    elif second_best_name == 'Logistic_Regression':
        best_model_instance = lr_pipeline
        best_model_name = second_best_name
        print(f"Выбрана модель: Logistic Regression Pipeline (lr_pipeline) как альтернатива.")
    elif second_best_name == 'Stacking_Classifier':
         if 'stacking_clf' in locals():
            best_model_instance = stacking_clf
            best_model_name = second_best_name
            print(f"Выбрана модель: Stacking Classifier (stacking_clf) как альтернатива.")
         else:
            print(f"ПРЕДУПРЕЖДЕНИЕ: Ни одна из подходящих моделей не была найдена. best_model_instance останется None.")
else:
    print(f"ПРЕДУПРЕЖДЕНИЕ: Неизвестное имя модели: {best_model_name}. best_model_instance останется None.")

# Проверяем, была ли успешно выбрана модель
if best_model_instance is None:
    print("\n--- ОШИБКА: Не удалось определить объект лучшей модели. Проверьте предыдущие шаги. ---")
    # Здесь можно добавить принудительную остановку выполнения ячейки, если это возможно в вашем окружении
    # raise RuntimeError("Не удалось выбрать модель для интерпретации.")
else:
    print(f"\n--- Объект лучшей модели ({best_model_name}) успешно определён для интерпретации. ---")
    print(f"Тип модели: {type(best_model_instance).__name__}")
```

```python
# --- Визуализация для лучшей модели ---
print(f"--- Создание визуализаций для лучшей модели: {best_model_name} ---")

# Убедимся, что best_model_instance определена
if best_model_instance is None:
    print("Невозможно создать визуализации: best_model_instance не определена.")
else:
    # --- Определение правильных предсказаний для CM ---
    y_pred_for_cm = None
    if best_model_name == 'Decision_Tree':
        y_pred_for_cm = y_pred_dt
    elif best_model_name == 'Random_Forest':
        y_pred_for_cm = y_pred_rf
    elif best_model_name == 'Gradient_Boosting':
        y_pred_for_cm = y_pred_gb
    elif best_model_name == 'Logistic_Regression':
        y_pred_for_cm = y_pred_lr
    elif best_model_name == 'Stacking_Classifier':
        if 'y_pred_stacking' in locals():
             y_pred_for_cm = y_pred_stacking
        else:
            print("Предсказания для Stacking не найдены. Пропускаем CM.")
    elif best_model_name == 'Dummy_Most_Frequent':
        y_pred_for_cm = y_pred_dummy
    else:
        print(f"Неизвестное имя модели для получения y_pred: {best_model_name}. Пропускаем CM.")

    # --- Получение вероятностей для ROC ---
    y_proba_best = None
    if best_model_name in ['Decision_Tree', 'Random_Forest', 'Gradient_Boosting', 'Logistic_Regression', 'Stacking_Classifier']:
        prob_map = {
            'Decision_Tree': y_proba_dt,
            'Random_Forest': y_proba_rf,
            'Gradient_Boosting': y_proba_gb,
            'Logistic_Regression': y_proba_lr,
            'Stacking_Classifier': y_proba_stacking if 'y_proba_stacking' in locals() else None
        }
        y_proba_best = prob_map.get(best_model_name)

    # --- Построение графиков ---
    fig, axes = plt.subplots(1, 2, figsize=(14, 6)) # Увеличили ширину и высоту

    # 1. ROC Curve
    if y_proba_best is not None:
        from sklearn.metrics import RocCurveDisplay
        RocCurveDisplay.from_predictions(y_test, y_proba_best, ax=axes[0])
        axes[0].set_title(f'ROC Curve - {best_model_name}', fontsize=14)
        axes[0].grid(True, linestyle='--', alpha=0.6) # Добавим сетку для наглядности
    else:
        axes[0].text(0.5, 0.5, f'ROC Curve\nNot Available\n({best_model_name})', 
                     horizontalalignment='center', verticalalignment='center', 
                     transform=axes[0].transAxes, fontsize=12, color='red')
        axes[0].set_title(f'ROC Curve - {best_model_name} (No Probabilities)', fontsize=14)
        axes[0].set_xlabel('False Positive Rate')
        axes[0].set_ylabel('True Positive Rate')

    # 2. Confusion Matrix
    if y_pred_for_cm is not None:
        cm = confusion_matrix(y_test, y_pred_for_cm)
        # Используем fmt='g', чтобы matplotlib сам решал, как форматировать числа (int или float), избегая научной нотации при необходимости
        sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', cbar_kws={'shrink': 0.8}, ax=axes[1])
        axes[1].set_title(f'Confusion Matrix - {best_model_name}', fontsize=14)
        axes[1].set_ylabel('True Label', fontsize=12)
        axes[1].set_xlabel('Predicted Label', fontsize=12)
        
        # Попробуем увеличить размер шрифта аннотаций, если матрица маленькая (2x2), иначе оставим по умолчанию
        if cm.shape == (2, 2):
            for text in axes[1].texts: # axes[1].texts содержит объекты текста аннотаций
                text.set_fontsize(14) # Устанавливаем размер шрифта для аннотаций
        
    else:
        axes[1].text(0.5, 0.5, f'Confusion Matrix\nNot Available\n({best_model_name})', 
                     horizontalalignment='center', verticalalignment='center', 
                     transform=axes[1].transAxes, fontsize=12, color='red')
        axes[1].set_title(f'Confusion Matrix - {best_model_name} (No Predictions)', fontsize=14)

    plt.tight_layout() # Улучшает расположение элементов
    plt.savefig('artifacts/figures/combined_plots_best_model.png', dpi=150, bbox_inches='tight') # Сохраняем с высоким разрешением и без лишних отступов
    plt.show()

    print("\n--- Комбинированные графики ROC и Confusion Matrix сохранены в artifacts/figures/combined_plots_best_model.png ---")
```
